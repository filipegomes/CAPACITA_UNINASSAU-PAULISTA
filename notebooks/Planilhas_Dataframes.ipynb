## Utils

Aqui são definidas algumas Funções para execuções de tarefas específicas dentro do contexto da manipulação das planilhas de Simulado.

```python
#remove points inside cells.
def remove_points(dataframe):
  '''
  dataframe: Pandas Dataframe

  Removes all points from incoming sheets and replaces for np.NaN values.
  '''
  heads = list(dataframe.keys())
  for head in heads:
    dataframe[head] = dataframe[head].replace('.', np.NaN)
  return dataframe

#clear function
def clear_data(dataframe, clear_all=True):
  '''
  dataframe: dataframe a ser inserido.
  clear_all: caso True, remove toda a linha que contenha pelo menos um NA. Em False, só removerá a linha se todos os valores forem NA.

  Essa função limpa os pontos e limpa linhas que contenham apenas valores NA.
  '''
  remove_points(dataframe)
  try:
    del dataframe['Total']
  except KeyError:
    pass
  
  if clear_all:
    dataframe = dataframe.dropna(thresh=4) #Precisam existir no mínimo 4 valores não NA.
  else:
    dataframe = dataframe.dropna(how='all')
  return dataframe

#check if all grade values are numeric and can be transformed as a Float, if not, make them np.NaN values.
def check_float(value):
  try:
    float(value)
    return value
  except ValueError:
    return np.NaN

    
def convert_areas_float(dataframe, area):
  '''
  dataframe: dataframe a ser inserido.
  area: dicionario contendo as áreas de conhecimento com "linguagens","natureza","humanas","matematica".

  Essa função transforma os dados das colunas de notas em Float.
  '''
  all_areas = area['linguagens']+area['natureza']+area['humanas']+ area['matematica']
  dataframe[all_areas] = dataframe[all_areas].astype(float)
  return dataframe 

def sum_areas(dataframe,area_conhecimento: dict,all_areas=True,areas=None):
  '''
  dataframe (pd.Dataframe()): Dataframe consolidade dos dias 1 e 2.
  area_conhecimento (dict): Dicionário com as áreas de conhecimento e respectivos códigos das disciplinas.
  all_areas (bool): Marcar como False se não forem tratadas todas as áreas de conhecimento.
  areas (list): Caso all_areas seja False, enviar uma lista nesse parâmetro com as áreas de conhecimento que serão utilizadas.
  '''

  if all_areas:
    dataframe['HUMANAS'] = dataframe[area_conhecimento['humanas']].sum(axis=1)
    dataframe['HUMANAS'] = dataframe['HUMANAS'].apply(lambda x: format(float(x),".4f"))

    dataframe['NATUREZA'] = dataframe[area_conhecimento['natureza']].sum(axis=1)
    dataframe['NATUREZA'] = dataframe['NATUREZA'].apply(lambda x: format(float(x),".4f"))

    dataframe['LINGUAGENS'] = dataframe[area_conhecimento['linguagens']].sum(axis=1)
    dataframe['LINGUAGENS'] = dataframe['LINGUAGENS'].apply(lambda x: format(float(x),".4f"))

    dataframe['MATEMATICA'] = dataframe[area_conhecimento['matematica']]
    dataframe['MATEMATICA'] = dataframe['MATEMATICA'].apply(lambda x: format(float(x),".4f"))
  else:
    for area in areas:
      if area.lower() != 'matematica':
        dataframe[area.upper()] = dataframe[area_conhecimento[area.lower()]].sum(axis=1)
        dataframe[area.upper()] = dataframe[area.upper()].apply(lambda x: format(float(x),".4f"))
      else:
        dataframe[area.upper()] = dataframe[area_conhecimento[area.lower()]]
        dataframe[area.upper()] = dataframe[area.upper()].apply(lambda x: format(float(x),".4f"))

  return dataframe

def create_output_data(dataframe,areas_de_conhecimento,data_codes,disciplina,cod_prova,etapa,turma, turno='Manhã'):
  '''
  dataframe (pd.Dataframe): Dataframe consolidado dos dias 1 e 2.
  areas_de_conhecimento (dict): Dicionário com as áreas de conhecimento e respectivos códigos das disciplinas.
  data_codes (pd.Dataframe): planilha com os dados de Códigos de Turma, Filial, Habilitação, Etc.
  disciplina (str): Disciplina onde será gerado o arquivo de importação.
  cod_prova (int): Qual prova será salva (qual a prova da Etapa? 1, 2, 3, 4?)
  etapa (int): Etapa atual para lançamento das notas.
  turma (str): Categoria da turma para consulta no dataframe de códigos.
  turno (str): Turno das turmas. Padrão é Manhã.
 
  '''

  dataframe_export = pd.DataFrame(columns=['CODCOLIGADA','CODCURSO','CODHABILITACAO','CODGRADE','TURNO','CODFILIAL','CODTIPOCURSO','RA','CODTURMA','CODPERLET','CODDISC','CODETAPA','TIPOETAPA','CODPROVA','CONCEITO','NOTA','NUMACERTOS'])
  dataframe_export['RA'] = pd.Series(list(dataframe['RA']))
  dataframe_export['CODCOLIGADA'] = 1
  dataframe_export['CODDISC'] = disciplina
  dataframe_export['CODPROVA'] = cod_prova
  dataframe_export['TIPOETAPA'] = 'N'
  dataframe_export['CODETAPA'] = etapa
  dataframe_export['TURNO'] = turno
  dataframe_export['CODTURMA'] = pd.Series(list(dataframe['Turma']))

  if disciplina in areas_de_conhecimento['humanas']:
    dataframe_export['NOTA'] = pd.Series(list(dataframe['HUMANAS'].str.replace('.', ',').replace('nan', np.NaN)))
  elif disciplina in areas_de_conhecimento['natureza']:
    dataframe_export['NOTA'] = pd.Series(list(dataframe['NATUREZA'].str.replace('.', ',').replace('nan', np.NaN)))
  elif disciplina in areas_de_conhecimento['linguagens']:
    dataframe_export['NOTA'] = pd.Series(list(dataframe['LINGUAGENS'].str.replace('.', ',').replace('nan', np.NaN)))
  else:
    dataframe_export['NOTA'] = pd.Series(list(dataframe['MATEMATICA'].str.replace('.', ',').replace('nan', np.NaN)))

  dataframe_export['CODCURSO'] = codigos.loc[turma,'CODCURSO']
  dataframe_export['CODHABILITACAO'] = codigos.loc[turma,'CODHABILITACAO']
  dataframe_export['CODGRADE'] = codigos.loc[turma,'CODGRADE']
  dataframe_export['CODFILIAL'] = codigos.loc[turma,'CODFILIAL']
  dataframe_export['CODTIPOCURSO'] = codigos.loc[turma,'CODTIPOCURSO']
  dataframe_export['CODPERLET'] = codigos.loc[turma,'CODPERLET']

  return dataframe_export
```

```python
#Loading each student as a dictionary with its RA as a Key.
def make_dict(list_dicts):
  students = {}
  keys = list((list_dicts[0].keys()))[1:]
  for student in list_dicts:
    students[format(student.get('RA'), '.0f')] = {key: student[key] for key in keys}
  for student in students.keys():
    if students[student].isnumeric():
      students.update(student=float(student))
  return students
```
